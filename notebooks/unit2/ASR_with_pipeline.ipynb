# ASR for english

```python
from datasets import load_dataset
from datasets import Audio
from transformers import pipeline
```

```python
minds = load_dataset("PolyAI/minds14", name="en-AU", split="train")
minds = minds.cast_column("audio", Audio(sampling_rate=16_000))
```

```python
asr = pipeline("automatic-speech-recognition")
```

```python
example = minds[0]
asr(example["audio"]["array"])
```

```python
example["english_transcription"]
```

#ASR for german

```python
minds_german = load_dataset("PolyAI/minds14", name="de-DE", split="train")
minds_german = minds_german.cast_column("audio", Audio(sampling_rate=16_000))
```

```python
example_german = minds_german[0]
example_german["transcription"]
```

```python
asr_german = pipeline("automatic-speech-recognition", model="maxidl/wav2vec2-large-xlsr-german")
asr_german(example_german["audio"]["array"])
```

# Bonus: what if we feed german to english asr

```python
asr(example_german["audio"]["array"])
```

```python
# and english to german asr

asr_german(example["audio"]["array"])
```
